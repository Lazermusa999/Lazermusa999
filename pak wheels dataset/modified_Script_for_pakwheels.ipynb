```python
from bs4 import BeautifulSoup
import csv
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore')
warnings.warn("ignore")
```

```python
url1 = 'https://www.pakwheels.com/used-cars/search/-/'
r = requests.get(url1)
soup = BeautifulSoup(r.content)

titles_dic1 = {}
titles_list = []
full_page = soup.find('div',class_='col-md-9 search-listing pull-right')
title = full_page.find_all('div',class_='search-title')
for t in title:
    name = t.a.h3.text
    titles_list.append(name)
    titles_dic1["Title"] = titles_list

price_dic1 = {}
price_list = []
prices = full_page.find_all('div',class_="price-details generic-dark-grey")
for p in prices:
    price = p.text.strip()
    price_list.append(price)
    price_dic1["Price"] = price_list
    
model_list = []
model_dic1 = {}
ul = full_page.find_all('ul',class_="list-unstyled search-vehicle-info-2 fs13")
for l in ul:
    li = l.find('li')
    model_list.append(li.text)
    model_dic1["Model"] = model_list
    
fuel_list = []
fuel_dic1 = {}
for l in ul:
    li = l.find_all('li')[2].text
    fuel_list.append(li)
    fuel_dic1["Fuel type"] = fuel_list
    
engine_list = []
engine_dic1 = {}
for l in ul:
    li = l.find_all('li')[3].text
    engine_list.append(li)
    engine_dic1["Engine capacity"] = engine_list
    
transmission_list = []
transmission_dic1= {}
for l in ul:
    li = l.find_all('li')[4].text
    transmission_list.append(li)
    transmission_dic1["Transmission type"] = transmission_list
    
loc_list = []
loc_dic1 = {}
ul_loc = full_page.find_all(class_='list-unstyled search-vehicle-info fs13')
for j in ul_loc:
    location = j.find('li')
    loc_list.append(location.text.strip())
    loc_dic1["Location"] = loc_list
```

```python
page_1 = {**titles_dic1,**price_dic1,**model_dic1,**fuel_dic1,**engine_dic1,**transmission_dic1}
df_1 = pd.DataFrame(page_1)
```

```python
dict_df={'Title':[],'Model':[],'Fuel':[],'Engine':[],'Transmission_type':[],'Location':[],'Price':[]}





for i in range(30):
    url = 'https://www.pakwheels.com/used-cars/search/-/?page='+str(i+1)
    r = requests.get(url)
    soup = BeautifulSoup(r.content)

    
    list_of_titles = []
    full_page = soup.find('div',class_='col-md-9 search-listing pull-right')
    title = full_page.find_all('div',class_='search-title')
    for t in title:
        name = t.a.h3.text
        list_of_titles.append(name)
        dict_df["Title"]+=list_of_titles

    price_list = []
    prices = full_page.find_all('div',class_="price-details generic-dark-grey")
    for p in prices:
        price = p.text.strip()
        price_list.append(price)
        dict_df["Price"]+=price_list
    
    model_list = []
    ul = full_page.find_all('ul',class_="list-unstyled search-vehicle-info-2 fs13")
    for l in ul:
        li = l.find('li')
        model_list.append(li.text)
        dict_df["Model"]+=model_list
    
    fuel_list = []

    for l in ul:
        li = l.find_all('li')[2].text
        fuel_list.append(li)
        dict_df["Fuel"]+=fuel_list
    
    engine_list = []
    for l in ul:
        li = l.find_all('li')[3].text
        engine_list.append(li)
        dict_df["Engine"]+=engine_list
    
    transmission_list = []
    for l in ul:
        li = l.find_all('li')[4].text
        transmission_list.append(li)
        dict_df["Transmission_type"]+=transmission_list
    
    loc_list = []
    ul_loc = full_page.find_all(class_='list-unstyled search-vehicle-info fs13')
    for j in ul_loc:
        location = j.find('li')
        loc_list.append(location.text.strip())
        dict_df["Location"]+=loc_list
    
    
```
